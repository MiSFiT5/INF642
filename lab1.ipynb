# Filter implementation

```python
from scipy.ndimage import median_filter
au_columns = [col for col in au_feature_1.columns if col.endswith('_r')]

window_sizes = [3,5,7,9]
dfs_filtered = []

for window_size in window_sizes:
    au_feature_1_filtered = pd.DataFrame()
    for au_column in au_columns:
        # median filter        
        au_feature_1_filtered[au_column] = median_filter(au_feature_1[au_column], window_size)
        au_feature_1_filtered.loc[(au_feature_1[" success"] == 0) | (au_feature_1[" confidence"]) < 0.5] = 0
        # au_feature_2[au_column + '_filtered'] = au_feature_2[au_column].rolling(window=window_size).median()
        # au_feature_3[au_column + '_filtered'] = au_feature_3[au_column].rolling(window=window_size).median()
    display(au_feature_1)
```
